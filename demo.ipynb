## 6. Experiment 2 — AI vs AI Benchmark

**Script:** `benchmark.py`

Runs fully automated AI-vs-AI games across all combinations of Black depth × White depth and reports win-rate matrix, nodes explored, pruning rate, effective branching factor (EBF), and average time per move. Results are saved to `results/benchmark/`.

Quick run (depths 1–2, 2 games per matchup):

```bash
python benchmark.py --depths 1 2 --games 2
```

Full reproduction run (depths 1–3, 3 games per matchup, 27 games total):

```bash
python benchmark.py --depths 1 2 3 --games 3
```

Analyze results:

```bash
python analyze.py --csv results/benchmark/<latest>.csv
```


---
## 7. Experiment 3 — Minimax vs Alpha-Beta

**Script:** `benchmark_algo.py`

Compares **plain Minimax** against **Alpha-Beta + TT + Killer + History** at the same search depth. Both sides search identically deep so the only variable is the algorithm. Results are saved to `results/benchmark_algo/`.

Key metrics reported:

| Metric | Description |
|---|---|
| **Node reduction** | % fewer nodes Alpha-Beta explores vs Minimax |
| **Pruning rate** | Fraction of subtrees cut by α-β; higher = better |
| **Speedup** | How many times faster Alpha-Beta is per move |
| **EBF** | Effective Branching Factor — ideal α-β approaches $\sqrt{b}$ of Minimax |

Quick run (depths 1–2, 2 games each):

```bash
python benchmark_algo.py --depths 1 2 --games 2
```

Full reproduction run (depths 1–3, 3 games each):

```bash
python benchmark_algo.py --depths 1 2 3 --games 3
```

Visualize results:

```bash
python analyze_algo.py results/benchmark_algo/<latest>.csv
```
